```python
# Check Stripplot with genderwise tip

sns.stripplot(data=t,x='sex',y='tip')
```

```python
# Check genderwise total_bill

sns.stripplot(data=t,x='sex',y='total_bill')
```

```python
sns.stripplot(data=t,x='time',y='total_bill',hue='sex')
```
