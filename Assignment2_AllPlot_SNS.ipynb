```python
# Diamond datasets

import seaborn as sns
```

```python
d = sns.load_dataset('diamonds')
d.head()
```

```python
# Histogram plot
sns.histplot(data=d)
```

```python
sns.histplot(data=d['carat'],bins=[0,0.5,1,1.5,2,2.5])
```

# Joint Plot

```python
sns.jointplot(data=d,x='depth',y='depth',kind='hex')
```

```python
sns.jointplot(data=d,x='depth',y='carat',kind='reg')
```

Scatter Plot

```python
sns.jointplot(data=d,x='x',y='y',hue='clarity')
```

```python
sns.jointplot(data=d,x='x',y='y',hue='clarity',kind='scatter')
```

KDE Plot

```python
sns.kdeplot(data=d)
```

Pair Plot

```python
sns.pairplot(data=d[['y','z']])
```

```python
sns.pairplot(data=d[['x','y','carat']],hue='carat')
```

```python
sns.pairplot(data=d)
```

Planet Dataset

```python
p = sns.load_dataset('planets')
p.head()
```

```python
sns.histplot(data=p['year'])
```

```python
# Joint Plot

sns.jointplot(data=p,x='distance',y='year')
```

```python
sns.jointplot(data=p,x='distance',y='mass',kind='hex')
```

```python
sns.jointplot(data=p,x='year',y='distance',kind='reg')
```

```python
# Scatter Plot

sns.jointplot(data=p,x='year',y='distance',hue='method')
```

```python
sns.kdeplot(data=p)
```

```python
p.orbital_period.value_counts()
```
