```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
 d = pd.read_csv('https://raw.githubusercontent.com/derekbanas/matplotlib/master/icecreamsales.csv')
 d.head(10)
```

```python
# Line Plot

plt.plot(d.Temperature,d.Sales,linestyle='--',color='green')
```

```python
# scatter Plot
plt.scatter(d.Temperature,d.Sales)
```

```python
# Bar/Barh plot

plt.bar(d.Temperature,d.Sales,width=0.6,color='green')
```

```python
# Horizontal plot
plt.barh(d.Temperature,d.Sales,alpha =0.3,color='red')
```

```python
# Pie Plot

s = d.Temperature.unique()
s
```

```python
plt.pie(s,labels=d.Temperature,autopct='%0.2f%%');
plt.show()
```

```python
# Histogram

plt.hist(d.Temperature)
```

```python
plt.hist(d.Sales)
```

```python
# Box plot

plt.boxplot(d.Sales,vert=False);
```
