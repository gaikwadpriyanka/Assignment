Apply all plots joint plot, pair plot, categoriwise Scatter plot,regression plot, hist plot, kde plot on diamonds dataset,tips & planet dataset

```python
import seaborn as sns
```

```python
# tips dataset

t = sns.load_dataset('tips')
t.head(5)
```

```python
# Histogram

sns.histplot(data=t)
```

```python
# Joint Plot

sns.jointplot(data=t,x='total_bill',y='tip')
```

```python
sns.jointplot(data=t,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=t,x='total_bill',y='tip',kind='reg')
```

```python
# Categoriwise Scatter Plot

sns.jointplot(data=t,x='total_bill',y='tip',hue='day')
```

```python
# kde plot

sns.kdeplot(data=t)
```

```python
# Pair Plot

sns.pairplot(data=t[['total_bill','tip']])
```

```python
sns.pairplot(data=t[['total_bill','tip','day']],hue='day')
```

```python
sns.pairplot(data=t,hue='day',palette='Reds')
```

```python
sns.boxplot(data=t)
```
