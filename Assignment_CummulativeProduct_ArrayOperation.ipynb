```python
# np.cumprod: cummulative product, row ,column,normal
```

```python
import numpy as np
```

```python
a = np.array([11,22,33,344,55])
```

```python
a
```

```python
np.cumprod(a)
```

```python
z = np.arange(30).reshape(6,5)
```

```python
z
```

```python
# Row wise Cumprod

np.cumprod(z,axis=0)
```

```python
# Columnwise cumprod

np.cumprod(z,axis=1)
```
